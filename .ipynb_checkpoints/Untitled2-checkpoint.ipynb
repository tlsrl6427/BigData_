```python
personality1 = pd.merge(left=rating, right=personality, how="inner", on='user_id')
personality1.rename(columns = {' movie_id':'movie_id',  ' emotional_stability' : 'neuroticism', ' openness':'openness',' agreeableness':'agreeableness',' conscientiousness':'conscientiousness',' extraversion':'extraversion'}, inplace=True)
personality1['agreeableness']=personality1['agreeableness'] * (5/7)
personality1['openness']=personality1['openness'] * (5/7)
personality1['neuroticism']=personality1['neuroticism'] * (5/7)
personality1['conscientiousness']=personality1['conscientiousness'] * (5/7)
personality1['extraversion']=personality1['extraversion'] * (5/7)
personality1['total'] = personality1['openness']+personality1['agreeableness']+personality1['neuroticism']+personality1['conscientiousness']+personality1['extraversion']
personality1.to_csv("personality1.csv")
```
